#Testes Estatísticos Básicos
Vamos ver alguns testes estatisticos com python. Vamos falar de testes de hipotéticos, significados estatísticos e usando Scipy para o teste T Student.

```python
#Nós usamos de muitas maneiras diferentes em datascience. Para isso, vamos relembrar nosso conhecimento de testes de hipotese. que é o cerne das atividades
#de análise por trás da experimentação. O objetivo dos testes de hipoteses é determinar se, por exemplo, duas diferentes condições que nós temos em um
#experimento resultaram em diferentes impactos.

#apenas importando como sempre
import numpy as np
import pandas as pd

#Agora vamos trazer algumas bibliotecas
from scipy import stats
```

```python
#Scipy é uma coleção interessante de bibliotecas, para ciencia de dados, você irá mais ou menos todas dessas bibliotecas. Ela inclui pandas e numpy,
#mas também bibliotecas de plotagem de gráficos, como matplotlib e um número de bibliotecas cientificas também.
```

```python
#Quando fazemos a testagem de hipoteses, nós temos duas demonstrações de interesse, a primeira é nossa atual explicação, que nós chamamos de hipotese
#alternativa, e o segundo é quando a explicação que temos não é o suficiente, e chamamos essa de hipotese nula. Nosso método atual é determinar se a
# hipotese nula é verdade ou não. Se encontrarmos que há uma diferença entre os grupos, então podemos rejeitar a hipotese nula e aceitamos nossa alternativa.
```

```python
#Vamos dar uma olhada em alguns dados de notas.
df = pd.read_csv('grades.csv', error_bad_lines = False)
df
```

```python
#Se percebeber, temos a nota de seis tarefas, vamos ver a sintese de algumas estatisticas desse dataframe
x = df.shape
print('Neste Dataframe temos %d linhas e %d colunas'%(x[0],x[1]))
```

```python
#Vamos segmentar esse dataframe em duas metades, aqueles que terminaram suas tarefas antes de do final de 2015, serão os adiantados e os que fizeram
#depois disso, serão os atrasados
early = df[pd.to_datetime(df['assignment1_submission']) < '2016']
early.head()
```

```python
#Agora que você já tem certa malicia com pandas, como vc faria para pegar os laters finisher? Faça uma tentativa.
late = df[~df.index.isin(early.index)] #O simbolo do '~' significa negação de um resultado, nesse caso. Os valores False
late.head()

```

```python
# Existem muitas outras maneiras de fazer isso, por exemplo, poderiamos apenas copiar e colar a primeira projeção, e mudar o sinal de '<' para '>='
#está okar, apenas se quiser mudar, lembrar de mudar nos dois lugares
#Poderiamos também fazer um join do dataframe dos adiantados com o df original, se fizer um 'left_join', você apenas mantera os itens no dataframe
#da esquerda, sendo entao também uma solução. Você também poderia fazer uma função que determinasse se alguém foi atrasado ou adiantado. E usar com
# a função '.apply()' e adicionar a um novo dataframe, está também é uma forma válida de tratar o problema. 
#Como foi dito, os elementos do tipo objeto dos pandas tem uma variedade de funções estatisticas associadas a ela.
#Se chamarmos a função 'mean()' diretamente ao dataframe. Veremos que calcula a média das tarefas. Vamos comparar a média das duas populações.

print(early['assignment1_grade'].mean())
print(late['assignment1_grade'].mean())
```

```python
# Certo, eles parecem similares. Mas, eles são o mesmos? O que dizemos com similar? é onde o teste t de student entra
#Isso nos permite formar diferentes hipoteses, ('eles sao diferentes'), como tão bem a hipotese nula, 
#("Eles são iguais') e então testa essa hipotese nula.
#Quando estiver testando as hipoteses, nós temos que escolher o nivel de significancia é um limite para 
#Qual a chance nós estamos dispostos a aceitar. Por exemplo, vamos usar um limite de 0.05 (5%) para um alpha;
#é geralmente usado numero, mas é arbitrário. 

# A biblioteca do Scipy contem um número de diferentes testes estatisticos e forma uma base para o teste de hipoteses
#iremos usar a função 'ttest_ind()' que faz um teste t independente ('ou seja, as populações não são
#relacionadas). o resultado desta função 'ttest_index()' são 't-statistic' e o 'p-value'.
#sendo este ultimo mencionado, a probabilidade, que é o mais importante para nós, e ela indica de 0 a 1
# (0 a 100%) das nossas hipoteses nulas serem verdade. Okay, vamos chamar o a função 'ttest_ind'.
from scipy.stats import ttest_ind
 
#Vamos correr nossa função com as duas populações da tarefa 1.

ttest_ind(early['assignment1_grade'],late['assignment1_grade'])


```

```python
#Então aqui vemos que a probabilidade é de 0.18, e é maior que o valor do nosso alfa
# de 0.05. Isso significa que não podemos rejeitar as hipoteses nulas. A hipotese nula era de que ambas populações fossem iguais
# e nós não temos certeza em nossa evidencia (porque é maior que o alfa) para chegar a uma conclusção contrária. Isso não significa que
#provamos que as populações são iguais.

```

```python
# por que não testamos as outras tarefas?
print(ttest_ind(early['assignment2_grade'],late['assignment2_grade']))
print(ttest_ind(early['assignment3_grade'],late['assignment3_grade']))
print(ttest_ind(early['assignment4_grade'],late['assignment4_grade']))
print(ttest_ind(early['assignment5_grade'],late['assignment6_grade']))
print(ttest_ind(early['assignment6_grade'],late['assignment6_grade']))
```

```python
#Ok, então parece que nesses dados nós não temos evidencias o bastante para sugerir que as populações diferem 
#em acordo as notas. Vamos dar uma olhada nesses valores p de probabilidade por um momento pois eles estão dizendo coisas
#que podem informar o design experimental futuramente. Por exemplo, uma das tarefas (tarefa 3) possui um valor p de 0.1.
# isso significa se nós aceitassemos um nivel de chance de 11%, isso poderia ter sido estatisticamente relevante. Como um pesquisador,
# isso poderia me sugerir que há algo ai que vale a pena ser considerada. Por exemplo, se nós tivessemos um numero pequeno de participantes
# (não temos), o se há algo unico sobre essa tarefa em especifico (qualquer que seja) então pode haver experimentos seguidos que poderiamos executar
```

```python
#Os valores p recentemente foram criticados por serem insuficientes para nos dizer o bastante sobre as interações
#que estão acontecendo, e duas outras tecnicas, intervalos de confiança e análise bayseana, são usados com mais regularidade.
#Um problema com os valores p é que conforme você rodas mais testes, mas possivelmente ira pegar um valor que é estatisticamente mais signicante
#apenas pela chances

#Vamos ver uma simulação disso. Primeiro, vamos criar um dataframe de 100 colunas, cada com 100 números. 
df1 = pd.DataFrame([np.random.random(100) for x in range(100)])
df1.head()
```

```python
# Pause e reflita, você entende a compreensão de lista e como foi que criei esse dataframe? 
#Você não precisa fazer uma compreensão de lista para fazer isso, mas deve saber como ler isso e imaginar como funcinona
# já que é uma das formas mais usadas nos foruns pela web.

# "df1 = pd.DataFrame([np.random.random(100) for x in range(100)])   #Antes do laço, é criado um df com 100 colunas de números randomicos
#  df1.head()"                                                       #ação em seguida repetida no laço 100x, criando um df 100x100
df1.shape
```

```python
#Okay, vamos criar um segundo dataframe
df2 = pd.DataFrame([np.random.random(100) for x in range(100)])
df2.head()
```

```python
# É esses dois DataFrames o mesmo? Talvez uma pergunta melhor seria,
# é uma coluna desse dataframe identica a outra??? 
#Vamos dar uma olhada, DIgamos que nosso valor critico é 0.1, ou um alpha de 10% e iremos comparar cada coluna
# no df1 a mesma coluna no df2. E então iremos avisar quando o valor de p é menor que 10% 
# O que significa que nós temos evidencias o suficiente para dizer que as colunas são diferentes.

#Vamos escrever isso em uma função chamada 'test_columns'
def test_columns(alpha = 0.1):
  num_diff = 0
  for col in df1.columns:
    #Podemos rodar 'ttest_ind()' entre as colunas dos dataframes
    teststat,pval = ttest_ind(df1[col],df2[col])
    #E checar o valor p e o alpha
    if pval <= alpha:
      #Agora apenas imprimimos e incrementamos a quantidade dos casos que satisfazem o if.
      print("Col {} é significavelmente estatisticamente diferente a alpha = {}, pval = {}".format(col,alpha,pval))
      num_diff += 1
    #E vamos então imprimir o resultado
  print('Total de números diferentes é {}, o que representa {}%'.format(num_diff,(num_diff)/len(df1.columns)*100))
#Chamando a função
test_columns()
```
