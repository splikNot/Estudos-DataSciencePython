```
def mothers_degree():
  dic = {'less than high school': 0,
         'high school': 0,
         'more than high school but not college':0,
         'college':0}
  df = pd.read_csv('NISPUF17.csv', index_col=0)
  for x in range((len(df))):
    temp = df['EDUC1'].iloc[x]
    if int(temp) == 1:
      dic['less than high school'] += 1
    elif int(temp) == 2:
      dic['high school'] += 1
    elif int(temp) == 3:
      dic['more than high school but not college'] += 1
    elif int(temp) == 4:
      dic['college'] += 1

  for x in dic:
    dic[x] = dic[x]/len(df)
  return dic

mothers_degree()

```

```
def average_influenza_doses():
  cbf = 0
  vacs = 0
  ncbf = 0
  nvacs = 0
  
  df = pd.read_csv('NISPUF17.csv')

  for x in range(len(df)):
    temp = df['CBF_01'].iloc[x]
    if temp == 1: #Se a criança foi amamentada
      cbf += 1 #Conta a criança
      if df['P_NUMFLU'].isnull().iloc[x] == True: #Se não sabe se a criança foi vacinada
        vacs += 0  #Não acrescenta na contagem
      else:
        vacs += int(df['P_NUMFLU'].iloc[x]) #Caso contrario, soma o numero de vacinas que a criança tomou
    elif temp == 2: #Se a criança não foi amamentada
      ncbf += 1  
      if df['P_NUMFLU'].isnull().iloc[x] == True:
        nvacs += 0
      else:
        nvacs += int(df['P_NUMFLU'].iloc[x])
  tup = (vacs/cbf, nvacs/ncbf) #Cria a tupla com as médias
  return tup

average_influenza_doses()
```

```
def vacined_child ():
  dic = {'male' : 0,
         'female' : 0}
  vac_contrac_m = 0
  vac_ncontrac_m = 0
  vac_contrac_f = 0
  vac_ncontrac_f = 0
  nvacs_m = 0
  nvacs_f = 0
  df = pd.read_csv('NISPUF17.csv', index_col = 0)

  for x in range(len(df)):
    if df['P_NUMVRC'].isnull().iloc[x] == False: #Se a criança foi vacinada
      temp = df[['SEX','HAD_CPOX']].iloc[x] #Verificar o sexo da criança
      if temp[0] == 1 and temp[1] == 1: #Se for um menino que teve catapora
        vac_contrac_m += 1
      elif temp[0] == 1 and temp[1] == 2: #Se for um menino, que não contraiu catapora
        vac_ncontrac_m += 1
      elif temp[0] == 2 and temp[1] == 1: #Se for uma menina, que contraiu a doença
        vac_contrac_f += 1
      elif temp[0] == 2 and temp[1] == 2: #Se for menina, que não contraiu catapora
        vac_ncontrac_f += 1 
    else:
      sex = df['SEX'].iloc[x]
      if sex == 1:
        nvacs_m += 1
      if sex == 2:
        nvacs_f += 1
      

  m = (vac_contrac_m/vac_ncontrac_m)
  f = (vac_contrac_f/vac_ncontrac_f)

  dic['male'] = m
  dic['female'] = f
  return dic

vacined_child()




```
